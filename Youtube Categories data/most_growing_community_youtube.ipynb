```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv('Meaningful_data.csv')#cleaned data 
df
```

```python
sns.barplot(x='Category',y='Growth Rate',data = df,color='blue')
plt.style.use('seaborn-v0_8-darkgrid')
plt.title('Community Growth in the year ')
plt.style.use('seaborn-v0_8-darkgrid')
plt.ylabel('Subscribers Growth (billions) between [2020 - 2024]')
plt.xlabel('Community Category')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
```
