```python
import numpy as np 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
df
```

```python
sns.lineplot(x='Month',y = 'PewDiePie (Millions)' , data = df,color = 'red',linestyle='--',label = 'PewDiepie')
sns.lineplot(x = 'Month',y = 'T-Series (Millions)',data = df,color = 'green',linestyle=':',label='T-Series')

plt.style.use('dark_background')
plt.xticks(rotation = 45)
plt.title('PewDipies Vs T-seires insane battle')
plt.xlabel('Months between [2018-2019]')
plt.ylabel('Pewdiepie Vs T-Series')
plt.tight_layout()
plt.grid(True)
plt.legend()
```
