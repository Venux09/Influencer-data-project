### 📊 Key Insights — PewDiePie vs T-Series Battle

1. The subscriber race between PewDiePie and T-Series began around **September 2018**, with both channels showing rapid growth.

2. By **December 2018**, the competition intensified as both channels reached nearly equal subscriber counts (~79 million), marking a critical midpoint in the battle.

3. In **March 2019**, T-Series overtook PewDiePie, gaining a lead of approximately **0.20 million subscribers**, signaling a major turning point in the competition.

4. In May 2019 T-series Won The battle with reaching 100 millions subs
